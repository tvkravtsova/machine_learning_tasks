1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Визначаємо симовлічну змінну
x = sp.symbols('x')

# Оголошуємо функцію
f = 10 * x - 5

# Обчислюємо похідну від f(x)
f_derivative = sp.diff(f, x)

# Перетворюємо функцію та її похідну в числові функції
f_func = sp.lambdify(x, f, 'numpy')
f_derivative_func = sp.lambdify(x, f_derivative, 'numpy')

# Задаємо числовий простір в діапазоні значень від 0 до 10
x_values = np.linspace(0, 10, 200)

# Обчислюємо значення функції f(x)
f_values = f_func(x_values)

# Для похідної f'(x) = 10, створюємо масив з тими ж розмірами
f_derivative_values = np.array([10] * len(x_values))

# Побудова графіків
plt.figure(figsize=(10, 6))

# Графік для f(x) = 10x - 5
plt.plot(x_values, f_values, label='f(x) = 10x - 5', color='blue')

# Графік для похідної f'(x) = 10
plt.plot(x_values, f_derivative_values, label="f'(x) = 10", color='red', linestyle='dashed')

plt.title("Графік функції f(x) та її похідної f'(x)")
plt.xlabel("Кількість годин (x)")
plt.ylabel("Оцінка / Похідна")
plt.legend()
plt.grid(True)

plt.show()


```

Висновок:
З графіка видно, що збільшення кількості годин навчання лінійно збільшує оцінку. Кожна додаткова година навчання робить рівномірний внесок у підвищення оцінки.
Швидкість зміни оцінки (похідна) є сталою і дорівнює 10 для будь-якого значення x.
Іншими словами, незалежно від того, скільки часу студент вже вивчав, кожна додаткова година навчання приносить йому однакову кількість додаткових балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Визначення змінної та функцій
x = sp.symbols('x')
f_sin = sp.sin(x)
f_ln = sp.ln(x)

# Обчислення похідних
df_sin = sp.diff(f_sin, x)
df_ln = sp.diff(f_ln, x)

# Перетворення символьних функцій у числові
f_sin_num = sp.lambdify(x, f_sin, "numpy")
df_sin_num = sp.lambdify(x, df_sin, "numpy")
f_ln_num = sp.lambdify(x, f_ln, "numpy")
df_ln_num = sp.lambdify(x, df_ln, "numpy")

# Задаємо числовий простір
x_vals_sin = np.linspace(-2 * np.pi, 2 * np.pi, 500)  # Для sin(x)
x_vals_ln = np.linspace(0.1, 10, 500)  # Для ln(x), уникаємо x=0

# Обчислюємо значення функцій та похідних
y_sin = f_sin_num(x_vals_sin)
y_sin_derivative = df_sin_num(x_vals_sin)
y_ln = f_ln_num(x_vals_ln)
y_ln_derivative = df_ln_num(x_vals_ln)

# Зображення графіків
plt.figure(figsize=(14, 6))

# Графік sin(x) та її похідної
plt.subplot(1, 2, 1)
plt.plot(x_vals_sin, y_sin, label="f(x) = sin(x)", color="blue")
plt.plot(x_vals_sin, y_sin_derivative, label="f'(x) = cos(x)", color="orange", linestyle="--")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Функція f(x) = sin(x) та її похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Графік ln(x) та її похідної
plt.subplot(1, 2, 2)
plt.plot(x_vals_ln, y_ln, label="f(x) = ln(x)", color="green")
plt.plot(x_vals_ln, y_ln_derivative, label="f'(x) = 1/x", color="red", linestyle="--")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Функція f(x) = ln(x) та її похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Графік функції sin(x) та її похідної cos(x):
Функція sin(x) періодична, з амплітудою від -1 до 1, повторюється через кожні 2𝜋.
У точках, де синусоїда досягає максимуму або мінімуму, косинусоїда перетинає вісь X. Це означає, що в цих точках швидкість зміни функції sin(x) дорівнює нулю (похідна дорівнює нулю).

У точках, де синусоїда перетинає вісь X, косинусоїда досягає екстремумів (максимуму або мінімуму). Це означає, що в цих точках швидкість зміни функції sin(x) максимальна за модулем.



Графік функції ln(x) та її похідної 1/x:

Функція ln(x) монотонно зростає, але зі зменшуваною швидкістю.

Чим більший x, тим менша похідна, що означає, що функція зростає все повільніше зі збільшенням x.




3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначення змінної та функції
x = sp.symbols('x')
f = x**2 - 5*x + 1

# Обчислення похідної
f_derivative = sp.diff(f, x)
f_derivative
```

```python
# Перетворення символьних функції і похідної у числові
f_numeric = sp.lambdify(x, f, 'numpy')
f_derivative_numeric = sp.lambdify(x, f_derivative, 'numpy')

# Визначення діапазону значень
x_values = np.linspace(0, 100, 200)
f_values = f_numeric(x_values)
f_derivative_values = f_derivative_numeric(x_values)

# Графіки функції і похідної
plt.figure(figsize=(10, 6))
plt.plot(x_values, f_values, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_values, f_derivative_values, label="f'(x) = 2x - 5", color='orange', linestyle='--')
plt.title('Графіки функції та її похідної')
plt.xlabel('Кількість вироблених одиниць (x)')
plt.ylabel('Загальні витрати від обсягу виробництва')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid()

plt.show()
```

Висновок:

У точці x=2.5 похідна рівна нулю, тобто на рівні виробництва 2.5 одиниць витрати не змінюються.

З огляду на те, що функція f(x) = x^2 - 5x + 1 параболічна і її похідна - лінійна функція, витрати на виробництво кожної наступної одиниці продукції рівномірно будуть зростати.
